```python
%matplotlib inline
import matplotlib.pyplot as plt

from numpy import arange, sin, pi, exp, log

def phi(k):
    return 1/(1+1j*k)**2
```

```python
ps = arange(-4, 4, 2/200)
phik_phikc = [(phi(k)*phi(-k)).real for k in ps]
plt.plot(ps, phik_phikc)

plt.xlabel('p', fontsize=18)
plt.ylabel('|$\phi$(p)|$^2$', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()
```

```python
ps = arange(-4, 4, 2/200)
phik_phikc = [k*(phi(k)*phi(-k)).real for k in ps]
plt.plot(ps, phik_phikc)

plt.xlabel('p', fontsize=18)
plt.ylabel('p|$\phi$(p)|$^2$', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()
```

```python
def A(k, N=1, alpha=1):
    return N/(k**2 + alpha**2)

N, alpha = 1, 1
delK = 2*alpha

ks = arange(-delK, delK, delK/200)
Ak2 = [A(k)**2 for k in ks]
plt.plot(ks, Ak2)
plt.bar(0, N**2/4, delK, fill=False, linewidth=2, edgecolor='red')
plt.xlabel('k', fontsize=18)
plt.ylabel('|A(k)|$^2$', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()
```

```python
def A(x, N=1, alpha=1):
    return N*pi*exp(-alpha*abs(x))/alpha

N, alpha = 1, 1
delX = log(4)/alpha

ks = arange(-2*delX, 2*delX, 2*delX/200)
Ak2 = [A(k)**2 for k in ks]

plt.plot(ks, Ak2)
plt.bar(0, A(0)**2/4, delX, fill=False, linewidth=2, edgecolor='red')

plt.xlabel('x', fontsize=18)
plt.ylabel('|$\psi$(x)|$^2$', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt


x = [0, 1, 0, 2, 6, 3, 13, 16, 9, 7, 2, 1]
px = np.asarray([xi/ogrenci_sayisi for xi in x])
puanlar = np.arange(5,61, 5)

plt.bar(puanlar, height=px, width=5, edgecolor='black')

plt.xlabel('g [alınan puan]', fontsize=18)
plt.ylabel('P(g) [olasılık]', fontsize=18)
plt.show()
```

```python
go = sum(px*puanlar) # g'nin ortalaması
go2 = sum(px*puanlar**2) # g^2'nin ortalaması
dg2 = go2-go**2
print(go, go2, dg2)
```

```python
def N(g, C=1):
    return C*exp(-(g-go)**2/dg2)

# N(g)'yi öğrenci sayısına normalize et
C = ogrenci_sayisi/sum(N(puanlar))

plt.plot(puanlar, N(puanlar))
```
