```python
%matplotlib inline
import matplotlib.pyplot as plt

from numpy import arange, sin, pi

def phi(k):
    return 1/(1+1j*k)**2
```

```python
ps = arange(-4, 4, 2/200)
phik_phikc = [(phi(k)*phi(-k)).real for k in ps]
plt.plot(ps, phik_phikc)

plt.xlabel('p', fontsize=18)
plt.ylabel('|$\phi$(p)|$^2$', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()
```

```python
ps = arange(-4, 4, 2/200)
phik_phikc = [k*(phi(k)*phi(-k)).real for k in ps]
plt.plot(ps, phik_phikc)

plt.xlabel('p', fontsize=18)
plt.ylabel('p|$\phi$(p)|$^2$', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()
```
