```python
%matplotlib inline
import matplotlib.pyplot as plt

from numpy import arange, sin, pi

def A(k, N=1, K=1):
    if -K <= k <= K:
        return N
    else:
        return 0
    
def psi(x, N=1, K=1):
    return 2*N*sin(K*x)/x
```

Consider a wave packet for which
$$
\begin{aligned}
A(k) &=N & &-K \leq k \leq K \\
&=0 & & \text { elsewhere }
\end{aligned}
$$
Calculate $\psi(x, 0)$, and use some reasonable definition of the width to show that (2-8) is satisfied.

```python
K, N = 2, 2
delX, delK = 2*pi/K, 2*K

ks = arange(-delK, delK, delK/200)
Ak = [A(k, K=K, N=N) for k in ks]
plt.plot(ks, Ak)

plt.xlabel('k', fontsize=18)
plt.ylabel('A(k)', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()
```

SOLUTION We have
$$
\psi(x, 0)=\int_{-K}^{K} d k N e^{i k x}=\frac{N}{i x}\left(e^{i K x}-e^{-K x}\right)=2 N \frac{\sin K x}{x}
$$
The definition of $A(k)$ easily shows that $\Delta k=2 K$. A reasonable definition of $\Delta x$ might be the distance between the two points at which $\psi(x)$ first vanishes as it gets away from $x=0$. This happens when $K x=\pm \pi$, so that $\Delta x=2 \pi / K$. It follows that
$$
\Delta k \Delta x=4 \pi
$$
which certainly satisfies (2-8).

```python
xs = arange(-1.5*delX, 1.5*delX, delX/200)

psix = [psi(x, K=K, N=N)  for x in xs]
plt.plot(xs, psix)

plt.scatter(-delX, 0, color='red')
plt.scatter(delX, 0, color='red')

plt.xlabel('x', fontsize=18)
plt.ylabel('$\psi$(x, 0)', fontsize=18)
plt.grid(zorder=-999, color='lightgray')
plt.show()

print(f'{delX = :.6f},  {delK = :.6f},  {delX*delK = :.6f}')
```
